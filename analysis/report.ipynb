The dataset contains 9358 entries. It has columns (Descriptions sourced from dataset location [here](https://archive.ics.uci.edu/ml/datasets/Air+Quality)): 

0. Date (DD/MM/YYYY)
1. Time (HH.MM.SS)
2. CO (Carbon Monoxide) Concentration in mg/m^3

4. True hourly averaged overall Non Metanic HydroCarbons concentration in microg/m^3 (reference analyzer)
5. True hourly averaged Benzene concentration in microg/m^3 (reference analyzer)
6. PT08.S2 (titania) hourly averaged sensor response (nominally NMHC targeted)
7. True hourly averaged NOx concentration in ppb (reference analyzer)
8. PT08.S3 (tungsten oxide) hourly averaged sensor response (nominally NOx targeted)
9. True hourly averaged NO2 concentration in microg/m^3 (reference analyzer)
10. PT08.S4 (tungsten oxide) hourly averaged sensor response (nominally NO2 targeted)
11. PT08.S5 (indium oxide) hourly averaged sensor response (nominally O3 targeted)
12. Temperature in Â°C
13. Relative Humidity (%)
14. AH Absolute Humidity

**Notes**

1. Despite several of the PT08 columns being aesthetically similar, they each refer to much different things. These are Tin Oxide (PT08.S1), Titania (PT08.S2), Tungsten Oxide (PT08.S3(NOx), PT08.S4(NO2)) and Indium Oxide (PT08.S5(O3)). These variables aren't measured in the same way that CO(GT), NMHC(GT), C6H6(GT), etc are measured but rather by an averaged sensor response. The parentheses at the end of PT08 likely refer to its related measured variable (Measured in micromg/m^3 or mg/m^3) If you use the two of them in conjunction, you would likely be able to find out how much of a particular type of PT08 is present.

2. Absolute humidity is often expressed in either grams/m^3 or grams per kilogram but the dataset creator does not specify which he uses.

3. Error values are displayed as -200.

4. CO is measured in mg/m^3. NMHCs, Benzene and NO2 are measured in microg/m^3. This difference in units means it is difficult to plot them all on the same graph while maintaining some semblance of comprehensibility.

## Research Question 1:
### Is the average pollution of this Italian City increasing over time or is it relatively constant?

For a question like this, some simple line charts will likely be able provide some insight. As the pollutants are classified in different units, we must split them up into several charts.

```python
#The simplest way to answer this question is to graph the change of pollutants over time. 
#We already have the graph of Carbon Monoxide over time, so we just need the graphs of the other pollutants 
plt.figure(figsize=(20, 6))
ax = sns.lineplot(data=df, x="Date", y="NMHC(GT)", label='NMHC(GT)', color="#3498db", ci=None)
sns.lineplot(data=df, x="Date", y="C6H6(GT)", label='C6H6(GT)', color='Orange', ci=None)
sns.lineplot(data=df, x="Date", y="NO2(GT)", label='NO2(GT)', color="#FE0000", ci=None)
ax.set_title('Pollutant Concentrations vs. Date')
ax.set_ylabel('micromg/m^3')

#Now look at the graphs of the pollutants as well as the original Carbon Monoxide graph.
plt.figure(figsize=(20, 6))
plot2 = sns.lineplot(data=df, x="Date", y="CO(GT)", color="Green", ci=None)
plot2.set_title('Carbon Monoxide Concentration vs. Date')
plot2.set_ylabel('mg/m^3')
```

If we compare start and end points for each of these lines, we can see in some cases that there is a definite upward trend. (Specifically Carbon Monoxide and NMHC(GT) have a noticeable upward trend, the other two have either a very slight upward or no specific trend)

What does this all mean? We can with certainty say that the pollution is increasing but this may only be a temporary event. As the data moves later into March and then April, temperatures are likely to increase as spring appears. It is also more likely that more individuals go out when temperatures do increase, so one potential cause could be an increase in tourists or the local population becoming more active.

We can also compare the temperature graph to the others in order to single out a likely explanation.

```python
plt.figure(figsize=(20, 6))
plot1 = sns.lineplot(data=df, x="Date", y="T", color="Black", ci=None)

plot1.set_title('Temperature (Black) & Carbon Monoxide Concentration (Green) vs. Date')
plot1.set_ylabel('Celsius')
plot2 = plot1.twinx()
sns.lineplot(data=df, x="Date", y="CO(GT)", color="Green", ax=plot2, ci=None)
plot2.set_ylabel('mg/m^3')
```

Despite some obvious differences, the temperature and Carbon Monoxide concentration graphs have similar patterns to one another, rising and falling in many of the same places. Is this just a coincidence or is there more to it? We can try thinking of a few reasons why these two may be so similar.

When the Temperature is hot, the Carbon Monoxide levels tend to rise as well, which could indicate that more people are active. To investigate this, we can compare the hydrocarbon levels to the temperature and see the results.

```python
plt.figure(figsize=(20, 6))
plot1 = sns.lineplot(data=df, x="Date", y="T", color="Black", ci=None)
plot1.set_title('Temperature (Black) & NMHC Concentration (Blue) vs. Date')
plot1.set_ylabel('Celsius')
plot2 = plot1.twinx()
sns.lineplot(data=df, x="Date", y="NMHC(GT)", color="#3498db", ax=plot2, ci=None)
plot2.set_ylabel('microg/m^3')
```

Again much like the Carbon Monoxide concentration, the Hydrocarbon levels tend to rise when temperature does, indicating it is likely that more people (Or devices that burn hydrocarbons and spew their emissions into the air) are more active on days where the temperature is higher. So does that mean that with a higher temperature comes higher amount of pollutants? We can, with confidence, say that this could be the case **for this dataset.** We cannot apply these observations to the real world since the dataset is not significant enough.

## Research Question 2: ##
### Does Temperature have an impact on the amount of pollution in the air? Does humidity? ###



This plays off of some of the findings we had in Research Question 1. Temperature seems to be related to the amount of pollution in the air. However, that doesn't account for humidity. 

```python
plt.figure(figsize=(20, 6))
rhplot = sns.histplot(data=df, x="Date", y="RH",color="Grey")
rhplot.set_title('Relative Humidity (%) & Temperature (Black) Vs. Time')

plot2 = rhplot.twinx()
sns.lineplot(data=df, x="Date", y="T", color="Black", ax=plot2, ci=None)
plot2.set_ylabel('Celsius')
```

This histogram plot shows the percentage amount of water content in the air. As we organize the data by **date**, there will be multiple time periods in a day that have a particular air water content. This is demonstrated by the coloration of the rectangles, the darker the coloration, the more common that particular relative humidity was during that day. 

```python
plt.figure(figsize=(20, 6))
rhplot = sns.histplot(data=df, x="Date", y="RH",color="Grey")
rhplot.set_title('Relative Humidity (%) & Carbon Monoxide concentration (Green) Vs. Time')

plot2 = rhplot.twinx()
sns.lineplot(data=df, x="Date", y="CO(GT)", color="lightGreen", ax=plot2, ci=None)
plot2.set_ylabel('mg/m^3')
```

Plotting Carbon Monoxide emissions against the Relative humidity does not provide much insight. In fact, it looks to be that relative humidity and the CO concentration are either related or have a weak relationship as there isn't a particular trend that both follow together.

We can try again using the other measurements.

```python
plt.figure(figsize=(20, 6))
rhplot = sns.histplot(data=df, x="Date", y="RH",color="Grey")

ax = rhplot.twinx()

sns.lineplot(data=df, x="Date", y="NMHC(GT)", label='NMHC(GT)', color="blue", ax=ax, ci=None)
sns.lineplot(data=df, x="Date", y="C6H6(GT)", label='C6H6(GT)', color='Orange', ax=ax, ci=None)
sns.lineplot(data=df, x="Date", y="NO2(GT)", label='NO2(GT)', color="#FE0000", ax=ax, ci=None)
ax.set_title('Relative Humidity (%) & Pollutant Concentration vs. Date')
ax.set_ylabel('micromg/m^3')
```

As before, there is not a clear relationship visible here. Hydrocarbon emissions spike on seemingly random days even when the average humidity tends low or high. It is impossible to determine a relationship between Benzene and humidity and a similar situation occurs when trying to compare Nitrogen Dioxide.

So what does this all mean? Temperature may have a positive relationship with Carbon Monoxide and Hydrocarbon levels but it doesn't appear to be the same with humidity. We can thus say that it is unlikely that Humidity shares a relationship with Pollution emission levels.
